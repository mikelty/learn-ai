```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
 
lfw_dataset = fetch_lfw_people(min_faces_per_person=50) 
# Downloads the 'Labeled faces in the wild dataset' from http://vis-www.cs.umass.edu/lfw/
#'Ariel Sharon', 'Colin Powell', 'Donald Rumsfeld', 'George W Bush',
#       'Gerhard Schroeder', 'Hugo Chavez', 'Jacques Chirac',
#       'Jean Chretien', 'John Ashcroft', 'Junichiro Koizumi',
#       'Serena Williams', 'Tony Blair'

_, h, w = lfw_dataset.images.shape
X = lfw_dataset.data
y = lfw_dataset.target
```

```python
def plot_tiles(images, h, w, rows=10, cols=5):
    plt.figure(figsize=(30,60))
    for i in range(rows * cols):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.xticks(())
        plt.yticks(())

U,s,VT = np.linalg.svd(X, full_matrices = True)
eigenfaces = [VT[i,:] for i in range(100)]

#rank1 = [U[:,[i]]@VT[[i],:] for i in range(100)]
#eigenfaces = [rank1[i][0,:] for i in range(100)]

plot_tiles(eigenfaces, h, w)
```
