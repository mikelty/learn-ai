```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# number of features
p = 2

# number of examples
n = 2000

# generate matrix of n (random) examples of p features with a column of all ones
X0 = np.random.rand(n,p)
onevec = np.ones(shape = (n,1))

X = np.concatenate((X0,onevec),axis=1)

# Classifier weights
w = [[-1], [2], [-0.4]]

# Multiply feature matrix with weights  yhat = X*w
yhat = X@w

# Decide which class based on whether yhat is >< 0
# sign function returns +1 when yhat(i)>0 and -1 when yhat(i)<0
pred_label = np.sign(yhat);

# Split data into categories for plotting

IndPlus1 = np.where(pred_label == 1)
IndMinus1 = np.where(pred_label == -1)

plt.scatter(X[IndMinus1[0],0],X[IndMinus1[0],1],color='r')
plt.scatter(X[IndPlus1[0],0],X[IndPlus1[0],1],color='b')
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.title("Case 2")
plt.axis('square')
plt.show()
```
